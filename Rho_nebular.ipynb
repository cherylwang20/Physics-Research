```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import scipy

#define distance to sun/star
a = np.linspace(0.1, 2.0, 100)

#define constants 
mu = 2.3
m_H = 1.67*10**(-27)
k_B = 1.38*10**(-23)
G = 6.67408 * 10**(-11)
#G = 4.3*10**(-3)
M_mars = 0.64171*10**(24)
M_sun = 1.989*10**(30)
R = 200.21
gamma = 1.369
autom = 1.496*10**(11)

#define parameters of Mars
P_bot = 10**5 
R_bot = 3.3895*10**6 #[m]
R_out = 11.1*10**3 + R_bot #[m]
#T_mars = 210 #K
#mu_mars = 43.34 

T_disk = 1000*(a/0.1)**(-3/7)
Cs_disk = np.sqrt(k_B*T_disk/mu/m_H) #CHECK UNITS [m/s^2]

#Cs_mars = np.sqrt(gamma*R*T_mars) 
#reference https://www.sciencedirect.com/science/article/pii/S0032063320302774#tbl2
A = G*M_mars/Cs_disk**2 #CHECK UNITS [m]

exp = np.exp(-A/R_out + A/R_bot)
rho_bot = P_bot/Cs_disk**2

rho_ne = rho_bot/exp*1000 #units of g/cm^3



K = np.sqrt(G*M_sun/((a*autom)**3)) #[1/s]
H = Cs_disk/K #[m]

Q = K*Cs_disk/G/H/rho_ne*1000
# what is rho_0????

hfont = {'fontname':'Helvetica'}

plt.plot(a, rho_ne,'-.', color = 'red')


plt.xlabel("Distance from Star (AU)",**hfont)
plt.ylabel("Density of Nebular (g/cm^3)",**hfont)


plt.rcdefaults()
plt.show()


plt.plot(a, Q)
plt.axhline(y = 1,color = 'green',linestyle = '--')
plt.xlabel("Distance from Star (AU)",**hfont)
plt.ylabel("Tooren Q",**hfont)
plt.yscale('log')
plt.show()


# minimum solar nebular, g/cm^2, compare to the rho_0. but less than Toomre Q. (refernece point)

# compare to the typical disk density at each location 
# Q = KCs_disk/GSigma; K= Omega
# Toomre Q, stable/unstable against self gravity
# ratio self gravity vs rotation & thermal (Sound speed)
# For disk to be stable have Q > 1; rotation in thermal has to dominate 
```
